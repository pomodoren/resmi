```python
import random
import datetime
import pandas as pd

# List for dates and values
datelist = []
valuelist = []

# Start date
d = datetime.date(2024, 1, 1)
# Generate dates for the year 2024, skipping Saturdays and Sundays
while d.year == 2024:
    if d.weekday() < 5:  # This checks for weekdays (Mon-Fri)
        datelist.append(pd.Timestamp(d))
        # Values will be appended later, only for weekdays
    d += datetime.timedelta(days=1)
    
# Generate values only for weekdays
valuelist = random.choices([0, 1, 2, 3], [5, 2, 1, 2], k=len(datelist))

# Create the DataFrame
df = pd.DataFrame(data={'Date': datelist, 'Value': valuelist})
df.set_index('Date', inplace=True)

df
```

```python
from calendarplot import create_year_calendar

hlmap = {
 0: ('gainsboro', None),
 1: ('lightblue', 'Roads'),
 2: ('tomato', 'Cycling'),
 3: ('teal', 'Walking'),
}

# create the calendar using the Value column
create_year_calendar(df['Value'], 2024, '2024 Calendar - Infra Sector Audits', hlmap=hlmap)

print()
print()
```
